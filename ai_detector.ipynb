```python
from datasets import load_dataset
import sys
import numpy as np
from IPython.display import display
```

```python
dataset=load_dataset("ComplexDataLab/OpenFake",split='train',streaming=True)
```

```python
data=dataset.take(2)
data=list(data)
data[1]['image']
data[1]
```

```python
image=data[1]['image']
size_of_image=image.size
print(size_of_image)
size_in_bytes=sys.getsizeof(image.tobytes())
size_in_mb=size_in_bytes/(1024*1024)
print(size_in_mb)
total_size=0
for key,value in data[1].items():
    if key=='image':
        size=sys.getsizeof(value.tobytes())

    else:
        size=sys.getsizeof(value)
    total_size+=size
print("Total size of one sample in MB:",total_size/(1024*1024),"MB")
```
